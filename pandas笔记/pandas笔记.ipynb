```python
stock = ["股票{}".format(i) for i in range(10)]
pandas.DataFrame(stock_change,index=stock)
```

## HDF5文件读写
- hdf5文件读取和存储时需要指定一个键，值为要存储的DataFrame文件
- hdf5存储三维数据
- key1 dataframe1二维数据
- key2 dataframe2二维数据

```python
day_open=pandas.read_hdf("day_open.h5")
```

```python
day_open
```

```python
day_open.to_hdf("test.h5",key="open")
```

```python
pandas.read_hdf("test.h5",key="open")
```
