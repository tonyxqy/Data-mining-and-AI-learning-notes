### Python默认的递归深度是有限制的，当递归深度超过默认值的时候，就会引发RuntimeError。理论在997
- 从Ackermann函数的定义中可以看出，Ackermann函数可以看成关于n的一个函数序列，其中第0个函数返回n+1，而第m个函数则是将第m-1个函数对1迭代n+1遍。对较小的m，该函数为：
- Ackermann(0,n)=n+1
- Ackermann(1,n)=n+2
- Ackermann(2,n)=2*n+3
- Ackermann(3,n)=2^(n+3)-3
- Ackermann(4,n)=2^2^2^……^2-3，乘幂中共有n+3个2。
- 当m≥4，Ackermann函数的增长快得惊人。Ackermann(4,0)=13，Ackermann(4,1)=65533，Ackermann(4,2)=2^65536-3有19729位，而Ackermann(4,3)则即使是位数也不易估计。Ackermann(5,0)=65533，

```python
class Time():
    def __init__(self,hour=0,minute=0,second=0):
        if type(hour) != int or type(minute) != int or type(second) != int:
            raise Exception('ALL ELEMENTS MUST BE NUMBERS')
        self.hour=hour
        self.minute=minute
        self.second=second
        self.timemodify()
    def timemodify(self):
        a = self.second//60
        self.second=self.second%60
        self.minute=self.minute+a
        b = self.minute//60
        self.minute=self.minute%60
        self.hour=self.hour+b
        self.hour=self.hour%24
    def __str__(self):
        return("时间是%d小时,%d分钟.%d秒" %(self.hour,self.minute,self.second))
    def __add__(self,n):
        second=self.second+n.second
        minute=self.minute+n.minute
        hour = self.hour+n.hour
        x=Time(hour,minute,second)
        x.timemodify()
        return x
    def time2int(self):
        return 59+55*60+10*60*60
    def increment(self,second):
        self.second=self.second+second
        self.timemodify()
```

```python
from random import randint
import pandas
#对单词后面出现词组的总数进行计算
def wordListSum(wordList):
    sum = 0
    for word, value in wordList.items():
        sum += value
    return sum
#随机取单词过程
def retrieveRandomWord(wordList):
    randIndex = randint(1, wordListSum(wordList))
    for word, value in wordList.items():
        randIndex -= value
        if randIndex <= 0:
            return word

def buildWordDict(text,n):
 # 剔除换行符和引号
    text = text.replace("\n", " ");
    text = text.replace("\"", "");
    punctuation = [',', '.', ';',':']
    for symbol in punctuation:
        text = text.replace(symbol, " "+symbol+" ");
    words = text.split(" ")
 # 过滤空单词
    words = [word for word in words if word != ""]
    wordDict = {}
    for i in range(1, len(words)):
        for j in range(1,n+1):
            if words[i-j] not in wordDict:
            #为单词新建一个词典
                wordDict[words[i-j]] = {}
            if words[i] not in wordDict[words[i-j]]:
                wordDict[words[i-j]][words[i]] = 0
            wordDict[words[i-j]][words[i]] = wordDict[words[i-1]][words[i]] + 1
    return wordDict

with open('whitefang.txt') as f:
    m=2
    senter=0
    text = str(f.read())
    wordDict = buildWordDict(text,1)
    df = pandas.DataFrame(wordDict)
    answer = ""
    currentWord = "I"
    while True:
        if currentWord != '.':
            answer += currentWord+" "
        currentWord = retrieveRandomWord(wordDict[currentWord])
        if currentWord=='.':
            senter+=1
            answer += currentWord+" "
            answer=answer.capitalize()
            print(answer,end="")
            answer=''
            if senter== m:
                 break
```
