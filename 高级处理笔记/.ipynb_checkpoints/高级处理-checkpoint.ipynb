## 缺失值处理

### 两种思路
1. 删除含有缺失值的样本
2. 替换/插补
#### 如何处理Nan
- 判断数据中是否存在Nan
    - pandas.isnull(df)    - 存在缺失值的地方返回True，其余返回False
    - pandas.notnull(df)
    
        - numpy.any(pandas.isnull(DataFrame))   -True说明Data中有True出现过说明有Nan
        - numpy.all(pandas.notnull(movie))-与上面相反
        - pandas.isnull(DataFrame名).any() -返回Series序列，哪一列存在Nan那一列的column对应的布尔值就会使True
        - pandas.notnull(movie).all()
- 删除含有缺失值的样本
    - DataFrame名.dropna(inplace=True/False)  - True原地删除，False返回新的
- 替换/插补
    - DataFrame名.fillna(value,inplace = )
    - DataFrame名["有Nan的那一列名"].fillna(DataFrame名["有Nan的那一列名"].mean(),inplace=True)

## 不是确实值nan，有默认标记的
    - 替换 ？-> np.nan
    - 处理np.nan缺失值的步骤
    

```python
# 读取数据
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
name = ["Sample code number", "Clump Thickness", "Uniformity of Cell Size", "Uniformity of Cell Shape", "Marginal Adhesion", "Single Epithelial Cell Size", "Bare Nuclei", "Bland Chromatin", "Normal Nucleoli", "Mitoses", "Class"]
data = pandas.read_csv(path, names=name)
```

```python
data.head()
```

```python
data_new=data.replace(to_replace='?',value=numpy.nan)
```

```python
data_new
```

```python
pandas.isnull(data_new).any()
```

```python
data_new["Bare Nuclei"].fillna(data_new["Bare Nuclei"].mean,inplace=True)
```

```python
pandas.isnull(data_new).any()
```

## 数据离散化
- one-hot编码&哑变量
### 如何实现数据的离散化
1. 分组  
    - 自动分组Series名=pandas.qcut(data,bins)
    - 自定义分组Series名=pandas.cut(data,[])
2. 将分组号的结果转换为one-hot编码
    - pandas.get_dumies(Series,prefix=)

```python
# 1）准备数据
data = pandas.Series([165,174,160,180,159,163,192,184], index=['No1:165', 'No2:174','No3:160', 'No4:180', 'No5:159', 'No6:163', 'No7:192', 'No8:184']) 
```

```python
data
```

```python
#2)分组
#自动分组
ar=pandas.qcut(data,3)
```

```python
ar
```

```python
pandas.get_dummies(ar,prefix="height")
```

```python
bins = [150, 165, 180, 195]
sr = pandas.cut(data, bins)
sr
```

```python
pandas.get_dummies(sr,prefix="height") 
```

### 股票离散化案例

```python
data = pandas.read_csv("stock_day.csv")
p_change = data["p_change"]
```

```python
p_change.head()
```

```python
#分组
ar= pandas.qcut(p_change,10)
ar
```

```python
pandas.get_dummies(ar,prefix="涨跌幅")
```

```python
ar.value_counts()
```

```python
#自定义分组
bins = [-100,-7,-5,-3,0,3,5,7,100]
sr = pandas.cut(p_change,bins)
sr
```

```python
data_new=pandas.get_dummies(sr,prefix="涨跌幅")
data_new.reset_index()
```

## 交叉表与透视表
探索两个元素之间的关系

```python
#探索星期数据以及涨跌幅数据是好是坏数据
stock = pandas.read_csv("stock_day.csv")
stock.index
```

```python
#pandas日期类型
date = pandas.to_datetime(stock.index)
stock["week"]=date.weekday
stock.head()
```

```python
import numpy
stock["pona"]=numpy.where(stock["p_change"]>0,1,0)
stock.head()
```

```python
asw = pandas.cut(stock["p_change"],[-100,0,100])
```

```python
stock["drop"]=pandas.get_dummies(asw,prefix="")["_(-100, 0]"]
stock["rise"]=pandas.get_dummies(asw,prefix="")["_(0, 100]"]
```

```python
stock
```

### 交叉表

```python
data = pandas.crosstab(stock["week"],stock["pona"])
data
```

```python
data.div(data.sum(axis=1),axis = 0).plot(kind="bar",stacked=True)
```

### 透视表


```python
stock.pivot_table(["pona"],index=["week"])
```

```python
data.div(data.sum(axis=1),axis = 0)
```

```python
stock.pivot_table(["pona"],index=["week"]).plot()
```
